```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as pit #saeborn is based on graph 
sns.set(color_codes=True) #add background to graph
%matplotlib.inline
#tellpyton to actually display the graph
```

```python
auto=pd.read_csv('Automobile.csv')
```

```python
auto.head()
```

```python
sns.displot(auto['highway_mpg'])
```

```python
sns.displot(auto['city_mpg'] ,kde=True,rug=True)#KDE Kernel density estimation 
```

```python
sns.jointplot(auto['engine_size'],auto['horsepower'])
```

```python
sns.jointplot(auto['engine_size'],auto['horsepower'],kind="hex")
```

```python
sns.jointplot(auto['engine_size'],auto['horsepower'],kind="kde")
```

```python
#pairwise plot. it create a matrix of axes and shows the relationship
#it also draws histogram of each variable on the diagonal axes.
sns.pairplot(auto[['normalized_losses','engine_size','horsepower']]) 
```

```python
#plotting with categorial data, adjust position using jitter
sns.stripplot(auto['fuel_type'],auto['horsepower'],jitter=True)
```

```python
sns.stripplot(auto['fuel_type'],auto['horsepower'],jitter=False)
```

```python
sns.swarmplot(auto['fuel_type'],auto['horsepower'])
```
